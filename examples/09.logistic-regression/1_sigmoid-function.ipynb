The sigmoid function, also known as the logistic function, is a function that maps any real number into a range between 0 and 1. The sigmoid function is defined as:

$\sigma(x) = \frac{1}{1 + e^{-x}}$

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate an array of x values
x = np.linspace(-10, 10, 100)

# Calculate the sigmoid values
y = sigmoid(x)

# Plot the sigmoid function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sigmoid Function', color='blue')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Sigmoid Function')
plt.axhline(y=0.5, color='gray', linestyle='--', label='Threshold (0.5)')
plt.legend()
plt.show()
```



In this example, we define the sigmoid function and apply it to a range of input values. The resulting plot shows how the sigmoid function maps any input value into the range between 0 and 1.
