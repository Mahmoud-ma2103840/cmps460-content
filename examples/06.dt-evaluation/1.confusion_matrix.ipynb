```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Example 1
y_true = ["Cat", "Not Cat", "Cat", "Cat",     "Cat", "Not Cat", "Not Cat", "Cat",     "Cat", "Not Cat"]
y_pred = ["Cat", "Cat",     "Cat", "Not Cat", "Cat", "Not Cat", "Cat",     "Not Cat", "Cat", "Cat"]
cm = confusion_matrix(y_true, y_pred, labels=["Cat", "Not Cat"])
print(cm)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Cat", "Not Cat"])
disp.plot()
plt.show()

cr = metrics.classification_report(y_true, y_pred)
print(f"Classification report:\n" f"{cr}\n")

# Example 2
y_true = [2, 0, 2, 2, 0, 1]
y_pred = [0, 0, 2, 2, 0, 2]
cm = confusion_matrix(y_true, y_pred)
print(cm)
```
