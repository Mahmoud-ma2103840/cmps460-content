## Gradient Descent

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import add_dummy_feature

plt.rc('font', size=12)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)
```

```python
import numpy as np

# Dataset
X = np.array([1.0, 1.2, 1.5, 1.7, 2.0, 2.2, 2.5, 2.7, 3.0, 3.5])
y = np.array([300, 350, 400, 425, 450, 475, 500, 525, 550, 600])

# Hyperparameters
learning_rate = 0.01
epochs = 1000

# Initialize parameters
w = 0.0
b = 0.0

# Gradient Descent
for epoch in range(epochs):
    # Predicted values
    y_pred = w * X + b
    
    # Mean squared error
    mse = np.mean((y_pred - y) ** 2)
    
    # Gradients
    dw = (2/len(X)) * np.dot(X, (y_pred - y))
    db = (2/len(X)) * np.sum(y_pred - y)
    
    # Update parameters
    w -= learning_rate * dw
    b -= learning_rate * db
    
    if epoch % 100 == 0:
        print(f'Epoch {epoch}: MSE = {mse}')

print("Optimal values:")
print("w =", w)
print("b =", b)
```

## Finding the minimum of f(x) = x^2

```python
def f(x):
    return(x*x)

# np.arange(start, stop, step) # start inclusive, stop exclusive
X = np.arange(-100, 100, 0.1)
y = [f(x) for x in X]

plt.figure(figsize=(7, 4))
plt.plot(X, y,  linewidth=2)
plt.xlabel(r"$x$")
plt.ylabel(r"$f\  (x)$")
plt.grid()
plt.show()
```

```python
def f_derivative(x):
    return 2*x

current_pos = (90, f(90))

learning_rate = 0.1
for i in range(10):
    new_x = current_pos[0] - learning_rate * f_derivative(current_pos[0])
    new_y = f(new_x)
    current_pos = (new_x, new_y)
    plt.plot(X, y)
    plt.scatter(current_pos[0], current_pos[1], color = "red")
    plt.pause(0.001)
    plt.clf()
```

```python
def f(x):
    return(x**2)

def f_grad(x):
    return(2*x)

def gd_1d(theta, n_epochs, eta):
    theta_path = [theta]
    for epoch in range(n_epochs):
        gradients = f_grad(theta)
        theta = theta - eta * gradients
        theta_path.append(theta)
    return theta_path
```

```python

eta = 0.2  # learning rate
n_epochs = 100
theta = 10

theta_path = gd_1d(theta,n_epochs,eta)

print("Initial theta\n",theta)
print("Final theta\n",theta_path[-1])
print("Value of f(x) at the 20th iteration\n",f(theta_path[20]))

y = [f(x) for x in X]

plt.plot(X, y , label='f(x)') 
plt.plot(theta_path, [f(x) for x in theta_path], 'o-', label='Gradient steps')
plt.legend()
plt.show()

plt.plot([i+1 for i in range(len(theta_path))], [f(x) for x in theta_path] , label='f(x)') 
plt.xlabel("Iteration")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

# Partial derivatives
## $f(x,y) = x^2 + 2y^2$
## $\dfrac{\partial }{\partial x} f(x,y) =  2x$  and  $\dfrac{\partial }{\partial y} f(x,y) =  4y$
$\begin{bmatrix} \dfrac{\partial }{\partial x} f(x,y) \\\dfrac{\partial }{\partial y} f(x,y)  \end{bmatrix} $
= $\begin{bmatrix} 2x \\ 4y \end{bmatrix} $
 

```python
#Calculate the value of the function at theta (x,y)
def f(theta):
    x = theta[0,0]
    y = theta[1,0]
    return(x**2 + 2*y**2)

#Calculate partial derivatives at theta (x,y)
def f_grad(theta):
    x = theta[0,0]
    y = theta[1,0]
    d_x = 2*x
    d_y = 4*y
    return (np.array([[d_x],[d_y]]))

def simple_gradient(theta, n_epochs, eta):
    theta_path = [theta]
    for epoch in range(n_epochs):
        gradients = f_grad(theta)
        theta = theta - eta * gradients
        theta_path.append(theta)

    return(theta_path)
```

```python
theta = np.array([[-5],[5]])
print("f at initial point",f(theta))
print("gradient of f at initial point",f_grad(theta).ravel()) 

lr = 0.01
max_iterations =100
theta_path = simple_gradient(theta, max_iterations, lr)

print("\n Gradient Descent with ", max_iterations, "Iterations and learning rate", lr)
print("Final point",theta_path[-1].ravel())
print("f at final point",f(theta_path[-1]))
print("gradient of f at final point",f_grad(theta_path[-1]).ravel()) 

#Change the max_iterations, since with 100 iterations the gradient is not zero; try more than 500 iterations
#We get a much smaller f(x,y)
lr = 0.01
max_iterations =1000
theta_path = simple_gradient(theta, max_iterations, lr)

print("\n Gradient Descent with ", max_iterations, "Iterations and learning rate", lr)
print("Final point",theta_path[-1].ravel())
print("f at final point",f(theta_path[-1]))
print("gradient of f at final point",f_grad(theta_path[-1]).ravel()) 
```

# Calculating the next steps for gradient descent
## Initial parameters  $(x,y) = (-5,5)$ and $\eta=0.01$

First we start at $(x,y) = (-5,5)$, calculate the partial derivatives at $(-5,5)$ and make a small step
in the opposite direction of to the new $x$.
$(x,y) = (-5,5)$


### Step 1: 
$\dfrac{\partial }{\partial x} f(-5,5) = 2*x =  2*(-5)  = -10$

$\dfrac{\partial }{\partial y} f(-5,5) = 4*y = 4*5      = 20$

$x = x - \eta \dfrac{\partial }{\partial x} f(-5,5)  = -5 - 0.01*(-10) = -4.9$
$y = y - \eta  \dfrac{\partial }{\partial y} f(-5,5) =  5 - 0.01*(20) =  4.8 $

### Step 2:
$\dfrac{\partial }{\partial x} f(-4.9, 4.8) = 2*(-4.9)  = -9.8$

$\dfrac{\partial }{\partial y} f(-4.9, 4.8) = 4*4.8     = 19.2$

$x = x - \eta \dfrac{\partial }{\partial x} f(-5,5)  = -4.9 - 0.01*(-9.8) = -4.802$

$y = y - \eta  \dfrac{\partial }{\partial y} f(-5,5) =  4.8 - 0.01*(19.2) =  4.608 $

and so on.

## Another example in 2D , two parameters
## $f(x,y) = sin(x)*cos(y)$
## $\dfrac{\partial }{\partial x} f(x,y) =  cos(y)*cos(x)$
## $\dfrac{\partial }{\partial y} f(x,y) =  -sin(x)*sin(y)$
$\begin{bmatrix} \dfrac{\partial }{\partial x} f(x,y) \\\dfrac{\partial }{\partial y} f(x,y)  \end{bmatrix} $
= $\begin{bmatrix} cos(y)*cos(x) \\ -sin(x)*sin(y) \end{bmatrix} $


```python
fig = plt.figure(figsize = (12,10))
ax = plt.axes(projection='3d')

x = np.arange(-5, 5.1, 0.2)
y = np.arange(-5, 5.1, 0.2)

X, Y = np.meshgrid(x, y)
Z = np.sin(X)*np.cos(Y)
print(Z.shape)
surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.cividis)
#surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.cividis, edgecolor='royalblue', lw=0.5, rstride=8, cstride=8, alpha=0.3)

# Set axes label
ax.set_xlabel('X', labelpad=20)
ax.set_ylabel('Y', labelpad=20)
ax.set_zlabel('Z=sin(x)*cos(x)', labelpad=20)
plt.show()
```

```python
def f(theta):
    x = theta[0,0]
    y = theta[1,0]
    return(np.sin(x)*np.cos(y))

def f_grad(theta):
    x = theta[0,0]
    y = theta[1,0]
    d_x = np.cos(y)*np.cos(x)
    d_y = -1*np.sin(x)*np.sin(y)
    return (np.array([[d_x],[d_y]]))
```
